# crawler THSR information

* Task:

```python
# setting each station name and their url
url = 'https://www.thsrc.com.tw/ArticleContent/2f940836-cedc-41ef-8e28-c2336ac8fe68'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
# save station name
index = []
stations = soup.find('div',id="nav-tab")
for station in stations:
    #print(station.string.strip())
    if((station.string.strip())==''):
        continue
    index.append(station.string.strip())
index
```

# Crawler GDP and CPI
Let's start to crawl GDP and CPI!
* Target web: 
    1. [Wiki GDP](https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal))  
    2. [Wiki CPI](https://en.wikipedia.org/wiki/Corruption_Perceptions_Index)
* Task:
    1. crawler GDP table (top 50: United States ~ New Zealand)
    2. organize into a DataFrame (columns: Country, Region, IMF_Estimate, IMF_Year, United_Nations_Estimate, United_Nations_Year, World_Bank_Estimate, World_Bank_Year)
    3. crawler CPI table which contains country and 2020 CPI (top 100: Denmark ~ Suriname)
    4. organize into a DataFrame (columns: Country, CPI_2020)
    5. merge GDP(DataFrame) and CPI(DataFrame), based on Country of GDP
    6. plot and text the names of GDP top 10 countrys

```python
# crawler GDP
GDP_dict = {'Country':[], 'Region':[], 'IMF_Estimate':[], 'IMF_Year':[],
                      'United_Nations_Estimate':[], 'United_Nations_Year':[], 
                      'World_Bank_Estimate':[], 'World_Bank_Year':[]}
keys = list(GDP_dict.keys())

url = 'https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
allfind = soup.find_all(['thead','tbody', 'tr','td'])

s=0#use to count top 50 country
for i,event in enumerate(allfind):
    if ((i>12)&(len(event.text)>40)&(s<50)):
        s=s+1
        eachlist=event.text.strip().split('\n')
        for i in range(len(eachlist)):
            GDP_dict[keys[i]].append(eachlist[i])  
```

```python
# create GDP DataFrame
GDP_DataFrame=pd.DataFrame(GDP_dict)
GDP_DataFrame
```

```python
# crawler CPI
url = 'https://en.wikipedia.org/wiki/Corruption_Perceptions_Index'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
allfind = soup.find_all(['thead','tbody', 'tr','td'])

CPI_dict = {'Country':[], 'CPI_2020':[]}

s=0#use to count top 100 country
for i,event in enumerate(allfind):
    if ((i>49)&(len(event.text)>40)&(s<179)):
        s=s+1
        alllist=event.text.strip().split('\n')
        CPI_dict['Country'].append(alllist[1].replace('\xa0',''))
        CPI_dict['CPI_2020'].append(int(alllist[2]))
```

```python
# CPI DataFrame
CPI_DataFrame=pd.DataFrame(CPI_dict)
CPI_DataFrame[0:100]
```

```python
# merge GDP and CPI DataFrame
GDP_DataFrame = GDP_DataFrame.merge(CPI_DataFrame, on='Country',how="left")
GDP_DataFrame
```

```python
plt.figure(figsize=(100, 150))
for i in range(len(GDP_DataFrame['World_Bank_Estimate'])):
    GDP_DataFrame['World_Bank_Estimate'][i] = int(re.sub(',', '',GDP_DataFrame['World_Bank_Estimate'][i]))
sns.relplot(x = GDP_DataFrame.CPI_2020, y = GDP_DataFrame.World_Bank_Estimate, hue = GDP_DataFrame.Region)
for i in range(10):
    plt.text(x=GDP_DataFrame['CPI_2020'][i], y=GDP_DataFrame['World_Bank_Estimate'][i], s=GDP_DataFrame['Country'][i])
```
