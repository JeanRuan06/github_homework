```python
import pandas as pd
df= pd.read_csv('churn_prediction_data/churn_prediction_data/train.csv')
##alter the categorial data to numerical type 
df_dummies = pd.get_dummies(df, columns=['Geography','Gender'])
df_dummies.head(5)
##alter some extreme number
minmax_col = ['CreditScore', 'Balance', 'EstimatedSalary']
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_dummies[minmax_col] = scaler.fit_transform(df_dummies[minmax_col])
df_dummies
```

```python
from sklearn.model_selection import train_test_split
all_inputs = df_dummies[['CreditScore','Age','Tenure','Balance','NumOfProducts','HasCrCard',
                 'IsActiveMember','EstimatedSalary','Geography_France','Geography_Germany','Geography_Spain',
                 'Gender_Female','Gender_Male']].values
all_classes = df_dummies['Exited'].values
(train_inputs, test_inputs, train_classes, test_classes) = train_test_split(all_inputs, all_classes, train_size=0.7, random_state=1,stratify=all_classes)
```

```python
##try to use dimension reduce
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(train_inputs)
train_inputs = pca.transform(train_inputs)
test_inputs = pca.transform(test_inputs)
#alter the number to elevate the predicted accuracy
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
train_inputs = sc.fit_transform(train_inputs)
test_inputs = sc.fit_transform(test_inputs)
```

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
from tensorflow import keras
model = keras.Sequential([
    keras.layers.Dense(32, input_shape=(13,), activation='relu'),
    keras.layers.Dense(13, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

# opt = keras.optimizers.Adam(learning_rate=0.01)

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(train_inputs,train_classes, epochs=100)
```

```python
y_hat = model.predict(test_inputs)
```

```python
y_hat = [0 if val<0.5 else 1 for val in y_hat]
accuracy_score(test_classes,y_hat)
```
